```python
# import the necessary packages
import pytesseract
import cv2

#https://github.com/tesseract-ocr/tesseract/blob/main/doc/tesseract.1.asc#languages-and-scripts (language and code)
# load the input image and convert it from BGR to RGB channel
x = "spa_1.jpg" 
# "german2.png" deu
# "chi2.JPG" chi_sim
# "spa_1.jpg" spa
# "japan.jpg" jpn
image = cv2.imread(x)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
text = pytesseract.image_to_string(rgb, lang='spa')

#display the image uploaded
from IPython.display import Image, display
display(Image(x))

# show the original OCR'd text
print("\nORIGINAL")
print("========")
print(text)
print("")
```

```python
from googletrans import Translator
translater = Translator()
out = translater.translate(text, dest="en")

# show the translated text
print("TRANSLATED")
print("==========")
print(out.text)
```
